### 1. What is the probability that the telescope will crash in the next T years?

```python
import random

my_lambda = 0.1
mu = 100
eta = 5 
nu = 100 
T = 15

simulations = 100000
crash_count = 0

for __ in range(simulations):
    print(f"{__+1} epochs")
    gyroscopes = [1] * 6 
    crashed = False

    for _ in range(T * nu):
        for i in range(len(gyroscopes)):
            if gyroscopes[i] == 1 and random.random() < my_lambda / nu:
                gyroscopes[i] = 0

        broken_count = gyroscopes.count(0)

        if broken_count >= 4:
            sleep_time = random.expovariate(mu)
            repair_time = random.expovariate(eta)
            remaining_gyroscopes = 6 - broken_count
            for _ in range(int((sleep_time+repair_time) * nu)):
                if remaining_gyroscopes <= 0:
                    crashed = True
                    break
                for i in range(len(gyroscopes)):
                    if gyroscopes[i] == 1 and random.random() < my_lambda / nu:
                        gyroscopes[i] = 0
                        remaining_gyroscopes -= 1

            if remaining_gyroscopes > 0:
                gyroscopes = [1] * 6 
            else:
                crashed = True
        if crashed:
            break
    if crashed:
        crash_count += 1

crash_probability = crash_count / simulations
print(f"The probability that the telescope will crash in the next {T} years: {crash_probability}")
```

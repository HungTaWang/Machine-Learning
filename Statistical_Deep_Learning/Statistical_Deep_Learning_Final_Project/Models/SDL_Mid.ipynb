### Loading data

```python
import numpy as np
from matplotlib import pyplot as plt
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.utils import plot_model
import pandas as pd

credit_card = pd.read_csv("creditcard.csv")

data = credit_card.to_numpy()

features = data[:, :-1]
target = data[:, -1]
```

#### Print features (in Numpy)

```python
print("Shape of features:", features.shape)
print(features)
```

#### Print target (in Numpy)

```python
print("Shape of target:", target.shape)
print(target)

fraud_counter = []
for i in range(len(target)):
    if(target[i]):
        fraud_counter.append(i)

print("Fraud_counter: ", fraud_counter)
print("Fraud number: ", len(fraud_counter))
```

### Data preprocessing

```python
train_features, val_features, train_targets, val_targets = train_test_split(features, target, test_size=0.3, random_state=65)

print("Original raw data:")
print(train_features[0])
print(train_targets[0])

scaler = StandardScaler()
train_features = scaler.fit_transform(train_features)
val_features = scaler.transform(val_features)

print("After standardized:")
print(train_features[0])

train_targets = keras.utils.to_categorical(train_targets)
val_targets = keras.utils.to_categorical(val_targets)

print("Target after one-hot encoding:")
print(train_targets[0])
```

### Build the model

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(train_features.shape[1],)),
    tf.keras.layers.Dense(1, activation='tanh'),
    tf.keras.layers.Dense(2, activation='tanh'),
    tf.keras.layers.Dense(train_targets.shape[1], activation='softmax')
])
model.summary()

print("\n Model structure:")
plot_model(model, to_file='model.png', show_shapes=True, show_layer_names=True)
```

### Experiment elaborating

```python
model.compile(
    optimizer=keras.optimizers.Adam(0.01),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
history = model.fit(
    train_features,
    train_targets,
    epochs=10,
    verbose=2,
    validation_data=(val_features, val_targets),
)
```

### Plot the graph

#### Accuracy

```python
plt.plot(history.history["accuracy"],label="Training Accuracy")
plt.plot(history.history["val_accuracy"],label="Validation Accuracy")
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```

#### Loss

```python
plt.plot(history.history["loss"],label="Training Loss")
plt.plot(history.history["val_loss"],label="Validation Loss")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```

### Calculate the gradient

#### Gradients of first-five epoch

```python
model.compile(
    optimizer=keras.optimizers.Adam(0.01),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
gradients = []

optimizer = keras.optimizers.Adam(0.01)
loss_fn = tf.keras.losses.CategoricalCrossentropy()

for epoch in range(5):
    with tf.GradientTape() as tape:
        predictions = model(train_features, training=True)
        loss_value = loss_fn(train_targets, predictions)

    grads = tape.gradient(loss_value, model.trainable_variables)
    optimizer.apply_gradients(zip(grads, model.trainable_variables))
    gradients.append(grads)

    for layer in model.layers:
        weights, biases = layer.get_weights()
        print(f"\nWeights and Biases at epoch {epoch + 1}:")
        print(f"\nLayer: {layer.name}")
        print(f"Weights shape: {weights.shape}")
        print(f"Weights: {weights}")
        print(f"Biases shape: {biases.shape}")
        print(f"Biases: {biases}")

for i, grads in enumerate(gradients):
    print(f"\nGradients at epoch {i + 1}:")
    for grad in grads:
        print(grad.numpy())
```

#### Final radient

```python
for layer in model.layers:
  weights, biases = layer.get_weights()
  print(f"\nLayer: {layer.name}")
  print(f"Weights shape: {weights.shape}")
  print(f"Weights: {weights}")
  print(f"Biases shape: {biases.shape}")
  print(f"Biases: {biases}")
```
